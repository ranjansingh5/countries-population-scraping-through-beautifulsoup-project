Import Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Scraping Logic

```python
countries_info = []

website = "https://www.worldometers.info/geography/countries-of-the-world/"

r = requests.get(website)
soup = BeautifulSoup(r.content, "html.parser")


rows = soup.find("table").find_all("tr")[1:]

for item in rows:
    cols = item.find_all("td")

    rank = cols[0].text.strip()
    country = cols[1].text.strip()
    population = cols[2].text.strip()
    region = cols[3].text.strip()

    all_countries = {
        "#": rank,
        "Country": country,
        "Population": population,
        "Region": region
    }

    countries_info.append(all_countries)
```

Create DataFrame and Save to CSV

```python
df = pd.DataFrame(countries_info)
df.to_csv("countries_population_2026.csv", index=False)

print("Scraping Completed Successfully!")
```
